## Scatter Plot

Create three scatter plots: the x axis for each plot is time and the y axis of the plot represents each of the three features.

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv(data_url)

plt.scatter(data['Time'], data['Outside Air Temperature'], s=6)
plt.xlabel('Time')
plt.ylabel('Outside Air Temperature \N{DEGREE SIGN}F')
plt.show()
```
